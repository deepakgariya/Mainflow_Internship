```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("C:/Users/deepa/Downloads/householdtask3.csv")
```

```python
data.head(10)
```

```python
#Scatter plot with year gainst own
plt.scatter(data['year'],data["own"])
            
#Adding Title to the Plot
plt.title("Scatter Plot")
            
#Setting the X and Y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Line chart with year against own
plt.plot(data['year'],data["own"])
            
#Adding Title
plt.title("Line Chart")
            
#Setting the X and Y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Bar Chart or Bar Plot
plt.bar(data['year'],data["own"])
            
#Adding Title
plt.title("Bar Chart or Bar Plot")
            
#Setting the X and Y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

#Adding Title
plt.title("Histogram")

#Showing the result
plt.show()
```
